k) Calculate the amount of money spent by Shapr day by day.

m) We would like to extend the analysis by looking at the marketing spend strategy by channel over time.

Calculate the amount spent per day per channel. We will store the result in a variable `df_facebook_channel_daily`.

**Use the argument as_index=False again to get a DataFrame as a result**

```python
df_facebook_channel_daily = df_facebook.groupby(['date','channel'], as_index = False)['spend'].sum().sort_values(by=['date'], ascending = True)

# group by channel & date and set index is False
# sort by date asc
```

```python
df_facebook_channel_daily
```

##### **Create a pivot table**

r) Make a pivot table with `date` as an index, channels as columns and `real_spend` as values.

Choose `sum` function for aggregate function.

We will store the result in a variable `df_facebook_pivot`.

Refer to your course for the correct syntax.

**Don't forget to apply the `.reset_index()` function to recreate an index**.


```python
data = df_facebook
values = ['real_spend']
index = ['date','channel']   # for better view, we set the index with date & channel
# columns = ['channel'] if we keep columns with channel, there will be 2 rows as headers
aggfunc = sum

df_facebook_pivot = pd.pivot_table(data = data, values = values, index = index, aggfunc =aggfunc).reset_index()
# we still need to define the parameter with variable

df_facebook_pivot
```

```python
data = df_facebook
values = ['real_spend']
index = ['date', 'channel']  # Set the index with date & channel
aggfunc = sum

# Group the data by date and channel, then create the pivot table
df_facebook_grouped = data.groupby(['date', 'channel'])['real_spend'].sum().reset_index()
df_facebook_pivot = pd.pivot_table(data=df_facebook_grouped, values=values, index=['date','channel'], aggfunc=aggfunc, fill_value=0)

# Display the pivot table
df_facebook_pivot
```
